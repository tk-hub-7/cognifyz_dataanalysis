```python
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
df=pd.read_csv("C:\\Users\\Karthik\\Downloads\\Dataset .csv")
df
```

```python
# Identify the restaurants with the highest and lowest number of votes.
```

```python
# Ensuring the 'Votes' column is numeric
df['Votes'] = pd.to_numeric(df['Votes'], errors='coerce')
```

```python
# Dropping rows with NaN votes if any exist
df_clean = df.dropna(subset=['Votes'])
df_clean
```

```python
# 1. Identifying the restaurant(s) with the highest number of votes
max_votes = df_clean['Votes'].max()
highest_voted_restaurants = df_clean[df_clean['Votes'] == max_votes][['Restaurant Name', 'Votes']]
highest_voted_restaurants
```

```python
# 2. Identifying the restaurant(s) with the lowest number of votes
min_votes = df_clean['Votes'].min()
lowest_voted_restaurants = df_clean[df_clean['Votes'] == min_votes][['Restaurant Name', 'Votes']]
lowest_voted_restaurants
```

```python
# Displaying Results
print(f"\n--- Highest Number of Votes ({max_votes}) ---")
print(highest_voted_restaurants)
print(f"\n--- Lowest Number of Votes ({min_votes}) ---")
print(f"\nTotal restaurants with {min_votes} votes: {len(lowest_voted_restaurants)}")
print("\nFirst 5 examples:")
print(lowest_voted_restaurants.head())
```

```python
# Analyze if there is a correlation between the number of votes and the rating of a restaurant.
```

```python
# Preprocessing: Ensuring 'Votes' and 'Aggregate rating' are numeric
# We use errors='coerce' to handle any non-numeric strings by turning them into NaN
df['Votes'] = pd.to_numeric(df['Votes'], errors='coerce')
df['Aggregate rating'] = pd.to_numeric(df['Aggregate rating'], errors='coerce')
```

```python
# Dropping rows with missing values in the columns needed for correlation
df_clean = df.dropna(subset=['Votes', 'Aggregate rating'])
df_clean
```

```python
# 1. Calculating the Pearson correlation coefficient
correlation = df_clean['Votes'].corr(df_clean['Aggregate rating'])
print(f"Correlation between Votes and Aggregate Rating: {correlation:.4f}")
```

```python
# 2. Visualizing the relationship with a Scatter Plot
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df_clean, x='Votes', y='Aggregate rating', alpha=0.5)
# Adding a title and labels
plt.title(f'Correlation Analysis: Votes vs Aggregate Rating\nPearson Correlation: {correlation:.4f}')
plt.xlabel('Number of Votes')
plt.ylabel('Aggregate Rating')
# Adding a grid for better readability
plt.grid(True, linestyle='--', alpha=0.6)

# Save the plot as an image
plt.savefig('votes_vs_rating_correlation.png')
```
