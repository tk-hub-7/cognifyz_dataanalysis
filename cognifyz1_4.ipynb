```python
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("C:\\Users\\Karthik\\Downloads\\Dataset .csv")
df
```

```python
df.info()
```

```python
# Determine the percentage of restaurants that offer online delivery.
df = df[df['Has Online delivery'].isin(['Yes', 'No'])]
df
```

```python
total_restaurants=len(df)
total_restaurants
```

```python
# Counting restaurants offering online delivery
online_delivery_counts = df['Has Online delivery'].value_counts()
online_delivery_counts
```

```python
# Calculating percentage
online_delivery_percentage = (online_delivery_counts / total_restaurants) * 100
online_delivery_percentage
```

```python
# Creating result table
result = pd.DataFrame({
    'Online Delivery': online_delivery_counts.index,
    'Restaurant Count': online_delivery_counts.values,
    'Percentage (%)': online_delivery_percentage.round(2).values
})

result
```

```python
# Comparing the average ratings of restaurants with and without online delivery.
```

```python
#converting aggregate rating to numeric(handle invalid values)
df['Aggregate rating']=pd.to_numeric(df['Aggregate rating'],errors='coerce')
# dropping rows with missing ratings
df=df.dropna(subset="Aggregate rating")
df
```

```python
# calculating average rating based on online delivery availability
average_rating_delivery=(
    df.groupby('Has Online delivery')['Aggregate rating']
    .mean()
    .round(2)
)
average_rating_delivery
```
