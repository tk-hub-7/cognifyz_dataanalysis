```python
import pandas as pd

# Load the dataset
df = pd.read_csv("C:\\Users\\Karthik\\Downloads\\Dataset .csv")

# Total number of restaurants
total_restaurants = len(df)

# Split multiple cuisines into individual entries
cuisines = df['Cuisines'].dropna().str.split(', ').explode()

# Count restaurants serving each cuisine
cuisine_counts = cuisines.value_counts()

# Select top 3 cuisines
top_cuisines = cuisine_counts.head(3)

# Calculate percentage of restaurants
percentage = (top_cuisines / total_restaurants) * 100

# Create final result
result = pd.DataFrame({
    'Cuisine': top_cuisines.index,
    'Restaurant Count': top_cuisines.values,
    'Percentage (%)': percentage.round(2).values
})

print(result)
```
