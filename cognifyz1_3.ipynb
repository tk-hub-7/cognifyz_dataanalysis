```python
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("C:\\Users\\Karthik\\Downloads\\Dataset .csv")
df
```

```python
df.info()
```

```python
# Creating a histogram or bar chart to visualize the distribution of price ranges among the restaurants.
```

```python
# counting restaurants by price range
price_count=df['Price range'].value_counts().sort_index()
price_count
```

```python
# creating a bar chart
plt.figure()
# plt.figure creates a new blank figure and ensures this plot does not overlap with other plot
plt.bar(price_count.index,price_count.values)
plt.xlabel("Price Range")     #labels the x-axis
plt.ylabel("Number of Restaurants")  #labels the y-axis
plt.title("Distribution of price ranges among restaurants") #adding title to the bar brapgh
plt.show() #displays the bargraph on screen
```

```python
# creating a histogram
plt.figure()
plt.hist(df['Price range'].dropna(),bins=4)
# selects price range column and removes all null values in that column that prevents errors and incorrect counts
# plt.hist() creates a histogram,bins=4 divides the price range into 4 groups 
plt.xlabel("Price range")   #labels the x-axis
plt.ylabel("Number of restaurants")  #labels the y-axis
plt.title("Histogram of restaurant price ranges") #adding title to the histogram
plt.show()  #displays the histogram on screen
```

```python
# Calculate the percentage of restaurants in each price range category.
total_restaurants=len(df)
total_restaurants
```

```python
percenta=(price_count/total_restaurants)*100
percenta
```

```python
result_df=pd.DataFrame({
    "Price range":price_count.index,
    "Restaurant count":price_count.values,
    "Percentage":percenta.round(2).values
     })
result_df
```
