```python
import pandas as pd
df=pd.read_csv("C:\\Users\\Karthik\\Downloads\\Dataset .csv")
df
```

```python
# 1.Identify the city with the highest number of restaurants in the dataset.
```

```python
total_restaurants=len(df)
total_restaurants
```

```python
# Counting number of restaurants per city
city_count=df['City'].value_counts()
city_count
```

```python
# finding the City with the highest number of restaurants
top_city = city_count.idxmax()
# idxmax() gives the city name with the highest count
top_city
```

```python
top_city_count = city_count.max()
# max() gives the number of restaurants
top_city_count
```

```python
print("City with the highest number of restaurants:")
print(f"{top_city} ({top_city_count} restaurants)")
```

```python
# Calculating the average rating for restaurants in each city.
```

```python
# Converting Aggregate rating to numeric (invalid values become NaN)
df['Aggregate rating'] = pd.to_numeric(df['Aggregate rating'], errors='coerce')
# errors='coerce' → converts invalid strings to NaN
df['Aggregate rating']

```

```python
# Dropping rows where rating is missing
df = df.dropna(subset=['Aggregate rating'])
# dropna() → removes rows with invalid ratings
df
```

```python
# Calculating average rating for each city
average_rating_city = (
    df.groupby('City')['Aggregate rating']
    .mean()
    .round(2)
    .sort_values(ascending=False)
)
average_rating_city
```

```python
print("Average rating of restaurants in each city:")
print(average_rating_city)
```

```python
# Determining the city with the highest average rating.
top_city=average_rating_city.idxmax()
top_city
```

```python
top_rating=round(average_rating_city.max(),2)
top_rating
```

```python
print("The city with the highest average rating:")
print(f"{top_city} is the highest rated city with a rating of {top_rating}")
```
